```python
import sys
from pathlib import Path

PROJECT_ROOT = Path.cwd().parent

if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

from core.knowledge import *
```

```python
topic = load_object("topic", "Mathematics")
```

```python
QUALITY_CHECKLIST = {

    "overview": True,

    "metadata": [

        "domain",
        "difficulty",
        "quality_level"

    ],

    "content": [

        "why_it_matters",

        "intuition",

        "key_concepts",

        "prerequisites",

        "historical_motivation",

        "timeline",

        "key_researchers",

        "applications",

        "open_problems",

        "future_directions",

        "faqs",

        "key_takeaways"

    ],

    "relationships": [

        "parent_topics",

        "child_topics",

        "related_topics"

    ]

}
```

```python
def evaluate_topic(topic):

    report = {}

    score = 0

    total = 0

    # ------------------------
    # Overview
    # ------------------------

    total += 1

    if topic.get("overview"):

        report["Overview"] = True

        score += 1

    else:

        report["Overview"] = False

    # ------------------------
    # Metadata
    # ------------------------

    for field in QUALITY_CHECKLIST["metadata"]:

        total += 1

        value = topic["metadata"].get(field)

        report[field] = bool(value)

        if value:

            score += 1

    # ------------------------
    # Content
    # ------------------------

    for field in QUALITY_CHECKLIST["content"]:

        total += 1

        value = topic["content"].get(field)

        report[field] = bool(value)

        if value:

            score += 1

    # ------------------------
    # Relationships
    # ------------------------

    for field in QUALITY_CHECKLIST["relationships"]:

        total += 1

        value = topic["relationships"].get(field)

        report[field] = True if value is not None else False

        if value is not None:

            score += 1

    percentage = round(score / total * 100)

    return report, percentage
```

```python
report, score = evaluate_topic(topic)

print("=" * 70)
print(topic["name"].upper())
print("=" * 70)

for key, value in report.items():

    symbol = "✅" if value else "❌"

    print(f"{symbol} {key}")

print()

print("=" * 70)
print(f"Knowledge Score : {score}%")
print("=" * 70)
```
