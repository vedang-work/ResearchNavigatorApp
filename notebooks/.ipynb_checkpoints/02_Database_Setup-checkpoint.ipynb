```python
conn = connect()

cursor = conn.cursor()

tables = [
    "topics",
    "researchers",
    "papers",
    "topic_links",
    "topic_researchers",
    "topic_papers"
]

for table in tables:

    print("\n" + "=" * 60)
    print(table.upper())
    print("=" * 60)

    cursor.execute(f"PRAGMA table_info({table})")

    for row in cursor.fetchall():
        print(dict(row))

close(conn)
```
